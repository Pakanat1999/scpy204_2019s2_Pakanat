```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('SQL.csv')
data
```

```python
z = data['z']
pg = data['petroMag_g']
zerr = data['zErr']
perr = data['petroMagErr_g']
```

```python
meanp = np.mean(pg)
meanp
```

```python
w = np.where(perr>20)
#w
```

The mean of petroMag_g is 21. I try to check the error value. So, there are many petroMagErr_g greater than 20. Therefore I got the graph like that.

```python
plt.plot(z,pg,'x')
plt.errorbar(z, pg, yerr=perr, xerr=zerr, fmt='.', ecolor='gray', elinewidth=0.1, capsize=0)
plt.xlim(0.3,0.4)
plt.ylim(-5,50)
plt.title('Z vs petroMag_g')
plt.xlabel('z')
plt.ylabel('petroMag_g')
```
